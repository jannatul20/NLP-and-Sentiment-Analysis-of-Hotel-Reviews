```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv("merged_dubai.csv")
df.head(5)
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer

# Feature extraction
tfidf = TfidfVectorizer(max_features=5000, lowercase=True, analyzer='word',
                        ngram_range=(1,1))
X = tfidf.fit_transform(df['Reviews'])
df2 = tfidf.vocabulary_
#Total words with their index in model :
print(df2)
```

```python
import csv

vocab = tfidf.vocabulary_

# Open a new CSV file for writing
with open('dubai_tfidf.csv', mode='w', newline='') as file:
    writer = csv.writer(file)

    # Write the header row
    writer.writerow(['Word', 'Index'])

    # Write the vocabulary data
    for word, index in vocab.items():
        writer.writerow([word, index])
```
